```python
#check data type

budgetdf.dtypes
```

```python
budgetdf = pd.read_csv('budget_data.csv', parse_dates=['Date'])
budgetdf['Date'] = budgetdf['Date'].dt.to_period(freq = 'M')
budgetdf.head()
```

```python
#check data type

budgetdf.dtypes
```

```python
#aggregate budget data

budgetdf = budgetdf.groupby(by=['Date']).sum()
budgetdf
```

```python
#summary stats

budgetdf.describe()
```

```python
#finding how many months we have
budgetdf.shape[0]
```

```python
#creating net change

# 1. take prior date value
# 2. present date value
# 3. minus 2 - 1

budgetdf['shift1'] = budgetdf['Profit/Losses'].shift(1)
budgetdf.head()
```

```python
#calculate monthly change

budgetdf['Monthy_change'] = budgetdf['Profit/Losses'] - budgetdf['shift1']
budgetdf.head(30)
```

```python
#get the max of monthly change

budgetdf['Monthy_change'].max()
```

```python
#getting the date/index of the greatest increase

budgetdf.Monthy_change[budgetdf.Monthy_change == budgetdf['Monthy_change'].max()].index.tolist()
max_date=budgetdf.Monthy_change[budgetdf.Monthy_change == budgetdf['Monthy_change'].max()].index.tolist()
```

```python
max_date[0].strftime('%b-%Y')
```

```python
#getting the date/index of the greatest decrease

budgetdf.Monthy_change[budgetdf.Monthy_change == budgetdf['Monthy_change'].min()].index.tolist()
min_date=budgetdf.Monthy_change[budgetdf.Monthy_change == budgetdf['Monthy_change'].min()].index.tolist()
```

```python
min_date[0].strftime('%b-%Y')
```

```python
#create a summary into a txt file


with open('Budget_Summary.txt',"w") as file:
    
    
    summary = f"""
    Financial Analysis
    -------------------------
    Total Months: {budgetdf.shape[0]}
    
    Total: ${round(budgetdf['Profit/Losses'].sum(),2)}
    
    Average Change: ${round(budgetdf['Monthy_change'].mean(),2)}
    
    Greatest Increase in Profits: {max_date[0].strftime('%b-%Y')} (${(budgetdf['Monthy_change'].max()):.0f})
  
    Greatest Decrease in Profits: {min_date[0].strftime('%b-%Y')} (${(budgetdf['Monthy_change'].min()):.0f})
    """
    
    file.write(summary)
```
